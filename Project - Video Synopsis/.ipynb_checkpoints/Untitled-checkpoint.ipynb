```python
import cv2
import numpy as np

video = cv2.VideoCapture('Video1.avi')
fps = video.get(cv2.CAP_PROP_FPS)

n_frame = video.get(cv2.CAP_PROP_FRAME_COUNT)
print(n_frame)
frames = []
while True:
    _, frame = video.read()
    if frame is None:
        break
    frames.append(frame)

video.release()

print("Video was Read!")

bg = np.median(frames[:200], axis=0).astype(dtype=np.uint8)

tubes = [f*0+bg for f in frames]
masks = []
start_frame = 0
tube_n = -1
for frame_number in range(len(frames)):

    frame = frames[frame_number]

    diff = cv2.cvtColor(cv2.absdiff(frame, bg), cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(diff, 60, 255, cv2.THRESH_BINARY)
    dilated = cv2.dilate(thresh, None, iterations=2)
    contours, rel = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    for i in range(len(contours)):

        (x, y, w, h) = cv2.boundingRect(contours[i])

        if cv2.contourArea(contours[i]) < 200:
            continue

        if y < 300 and y + h > 300 and y+h - 300 < 8 and y+h-300 > 1 and w > 40:
            (x, y, w, h) = (x-1, y-1, w+2, h+2)
            if rel[0][i][3] == -1:
                tube_n += 1
                old = initBB = ref = (x, y, w, h)
                tracker = cv2.TrackerMedianFlow_create()
                tracker.init(frame, initBB)
                d_c = 0
                coords = []
                for t_f in range(frame_number, 0, -1):
                    if initBB is not None:
                        fr_t = frames[t_f]
                        (success, box) = tracker.update(fr_t)

                        if success:
                            (x, y, w, h) = [int(v) for v in box]
                            dist = (x - old[0]) ^ 2 + (y - old[1]) ^ 2 + (x + w - old[0] - old[2]) ^ 2 + (
                                    y + h - old[1] - old[3]) ^ 2

                            coords.append((x, y, w, h))
                            start_frame = t_f
                            old = initBB = (x, y, w, h)
                            if dist == 0:
                                d_c += 1
                            else:
                                d_c = 0

                            if d_c == 10:
                                break

                (f_x, f_y, f_w, f_h) = (x, y, w, h)
                tracker = cv2.TrackerMedianFlow_create()
                tracker.init(frame, ref)
                d_c = 0
                for t_f in range(frame_number+1, len(frames), 1):
                    if initBB is not None:
                        fr_t = frames[t_f]
                        (success, box) = tracker.update(fr_t)

                        if success:
                            (x, y, w, h) = [int(v) for v in box]
                            dist = (x - old[0]) ^ 2 + (y - old[1]) ^ 2 + (x + w - old[0] - old[2]) ^ 2 + (
                                      y + h - old[1] - old[3]) ^ 2
                            if dist == 0:
                                d_c += 1
                            else:
                                d_c = 0

                            old = initBB = (x, y, w, h)
                            coords.insert(0, (x, y, w, h))

                            if d_c == 10:
                                break

                coords.reverse()
                start_pos = (f_x, f_y, f_w, f_h)
                masks.append((tube_n, start_frame, coords, start_pos))


for i in range(len(masks)):
    (tube_n, start_frame, coords, start_pos) = masks[i]
    (f_x, f_y, f_w, f_h) = start_pos
    for j in range(len(tubes)):
        current = tubes[j]
        if (current[f_y:f_y + f_h, f_x:f_x + f_w, :] == bg[f_y:f_y + f_h, f_x:f_x + f_w, :]).all():
            for c_n, coord in enumerate(coords):
                (x, y, w, h) = coord
                frame = frames[c_n+start_frame]
                window = frame*0
                window[y:y+h, x:x+w, :] = frame[y:y+h, x:x+w, :]
                time_t = np.round((start_frame+c_n)/fps, decimals=1)
                cv2.putText(window, str(time_t), (int((2*x+w)/2), int((2*y+h)/2)),
                            cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 255, 0))
                tubes[j+c_n][y:y+h, x:x+w, :] = frame[y:y+h, x:x+w, :]*0
                tubes[j+c_n] = cv2.add(tubes[j+c_n], window)
            break
```

```python
final_len = 0
for i in range(len(tubes)):
    f = tubes[i]
    if (f == bg).all():
        final_len = i
        break
    cv2.imshow("f", f)
    cv2.waitKey(40)

print(final_len)
cv2.destroyAllWindows()
```

```python
pathes = []
for i in range(len(masks)):
    (tube_n, start_frame, coords, start_pos) = masks[i]
    (f_x, f_y, f_w, f_h) = start_pos
    path = []
    for c_n, coord in enumerate(coords):
        (x, y, w, h) = coord
        path.append((int((2*x+w)/2), int((2*y+h)/2)))
    pathes.append(path)    
bg = np.median(frames[:200], axis=0).astype(dtype=np.uint8)
```

```python
b = bg*0
i = 0
X = []
for p in pathes:
    i = i + 1
    for p2 in p:
        (x,y) = p2
        if (y > 250):
            cv2.circle(b, p2, 3, (0,255,255), 3)
            X.append([x, y])

cv2.imshow("f", b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
